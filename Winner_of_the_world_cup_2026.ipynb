```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

reference_cols = [
    'home_elo','away_elo','elo_diff',
    'home_avg_overall','home_max_overall','home_avg_attack','home_avg_defense',
    'home_avg_pace','home_avg_shooting','home_avg_passing',
    'away_avg_overall','away_max_overall','away_avg_attack','away_avg_defense',
    'away_avg_pace','away_avg_shooting','away_avg_passing'
]

target_cols = ['home_avg_pace','home_avg_shooting','home_avg_passing',
               'away_avg_pace','away_avg_shooting','away_avg_passing']

before_counts = teams_match[target_cols].isnull().sum()

scaler = StandardScaler()
scaled_values = scaler.fit_transform(teams_match[reference_cols])
scaled_data = pd.DataFrame(scaled_values, columns=reference_cols, index=teams_match.index)

imputer = KNNImputer(n_neighbors=5, weights='distance')
imputed_scaled = imputer.fit_transform(scaled_data)

imputed_data = pd.DataFrame(imputed_scaled, columns=reference_cols, index=teams_match.index)
imputed_data[reference_cols] = scaler.inverse_transform(imputed_data[reference_cols])

teams_match[target_cols] = imputed_data[target_cols]

after_counts = teams_match[target_cols].isnull().sum()
print("Before :\n", before_counts)
print("\nAfter :\n", after_counts)
```

```python
data['result'] = np.select(
    [data['home_goals'] > data['away_goals'], data['home_goals'] == data['away_goals']],
    [2, 1],
    default=0
)

print(data[['home_goals','away_goals','result']].head(10))
```

```python
data["attack_diff"] = data["home_avg_attack"] - data["away_avg_attack"]

data["defense_diff"] = data["home_avg_defense"] - data["away_avg_defense"]

data["overall_diff"] = data["home_avg_overall"] - data["away_avg_overall"]

data["form_win_rate_diff"] = (
    data["home_form_win_rate"] - data["away_form_win_rate"])

data["form_scored_diff"] = (
    data["home_form_scored"] - data["away_form_scored"])

data["form_conceded_diff"] = (
    data["home_form_conceded"] - data["away_form_conceded"])

```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    log_loss
)

from xgboost import XGBClassifier


# 1. Remove Draw Matches


data_no_draw = data[data["result"] != 1].copy()



# Home Win = 1
# Away Win = 0

data_no_draw["result_binary"] = data_no_draw["result"].map({
    2: 1,
    0: 0
})


data_no_draw['_date'] = pd.to_datetime(data_no_draw['_date'])
data_recent = data_no_draw[
    (data_no_draw['_date'] >= '2017-01-01') &
    (data_no_draw['_date'] <= '2025-12-31')
].copy()

print("Number of rows after filtering:", data_recent.shape)
print("\nNew is_neutral ratio:\n", data_recent.groupby('is_neutral')['result_binary'].mean())
print("\nResults percentage:\n", data_recent['result_binary'].value_counts(normalize=True) * 100)


X = data_recent.drop(columns=["result","result_binary","home_goals","away_goals","_home_team","_away_team","_date","_tournament",
"is_neutral",'is_world_cup', 'is_continental',
])

y = data_recent["result_binary"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print("\nTrain shape:", X_train.shape)
print("Test shape:", X_test.shape)


xgb = XGBClassifier(objective="binary:logistic", eval_metric="logloss", random_state=42)

param_grid = {
    "n_estimators": [200, 300, 500, 700],
    "learning_rate": [0.01, 0.03, 0.05, 0.1],
    "max_depth": [3, 4, 5, 6],
    "min_child_weight": [1, 3, 5],
    "subsample": [0.7, 0.8, 0.9, 1.0],
    "colsample_bytree": [0.7, 0.8, 0.9, 1.0],
    "gamma": [0, 0.1, 0.2, 0.3]
}

search = RandomizedSearchCV(estimator=xgb,param_distributions=param_grid,n_iter=30,scoring="accuracy",cv=5,verbose=1,random_state=42,n_jobs=-1)

search.fit(X_train, y_train)

best_model = search.best_estimator_

print("=" * 50)
print("Best Parameters")
print(search.best_params_)
print("=" * 50)


preds = best_model.predict(X_test)
probs = best_model.predict_proba(X_test)

print("Accuracy :", accuracy_score(y_test, preds))
print("Log Loss :", log_loss(y_test, probs))
print(classification_report(y_test, preds))
print(confusion_matrix(y_test, preds))


importance = pd.DataFrame({"Feature": X.columns,"Importance": best_model.feature_importances_})

importance = importance.sort_values(by="Importance", ascending=False)

print("\nTop Features")
print(importance)


low_features = importance[importance["Importance"] < 0.005]

print("\nLow Importance Features")
print(low_features)
```

```python
def build_match_row(team_a, team_b, teams_data, X_columns, is_neutral=1, is_world_cup=1, is_continental=0):
    a_match = teams_data[teams_data['team'] == team_a]
    b_match = teams_data[teams_data['team'] == team_b]

    if a_match.empty:
        raise ValueError(f"الفريق '{team_a}' مش موجود في teams_data")
    if b_match.empty:
        raise ValueError(f"الفريق '{team_b}' مش موجود في teams_data")

    a = a_match.iloc[0]
    b = b_match.iloc[0]

    row = {
        'home_elo': a['elo'],
        'away_elo': b['elo'],
        'elo_diff': a['elo'] - b['elo'],

        'home_avg_overall': a['overall'],
        'home_max_overall': a['max_overall'],
        'home_avg_attack': a['attack'],
        'home_avg_defense': a['defense'],
        'home_avg_pace': a['pace'],
        'home_avg_shooting': a['shooting'],
        'home_avg_passing': a['passing'],

        'away_avg_overall': b['overall'],
        'away_max_overall': b['max_overall'],
        'away_avg_attack': b['attack'],
        'away_avg_defense': b['defense'],
        'away_avg_pace': b['pace'],
        'away_avg_shooting': b['shooting'],
        'away_avg_passing': b['passing'],

        'overall_diff': a['overall'] - b['overall'],
        'attack_diff': a['attack'] - b['attack'],
        'defense_diff': a['defense'] - b['defense'],

        'home_form_scored': a['form_scored'],
        'home_form_conceded': a['form_conceded'],
        'home_form_win_rate': a['form_win_rate'],
        'away_form_scored': b['form_scored'],
        'away_form_conceded': b['form_conceded'],
        'away_form_win_rate': b['form_win_rate'],

        'form_win_rate_diff': a['form_win_rate'] - b['form_win_rate'],
        'form_scored_diff': a['form_scored'] - b['form_scored'],
        'form_conceded_diff': a['form_conceded'] - b['form_conceded'],

        'is_neutral': is_neutral,
        'is_world_cup': is_world_cup,
        'is_continental': is_continental,
    }

    return pd.DataFrame([row])[X_columns]
```

```python
def simulate_knockout_match(team_a, team_b, teams_data, model, X_columns, deterministic=True):
    row_a_home = build_match_row(team_a, team_b, teams_data, X_columns)
    probs_a_home = model.predict_proba(row_a_home)[0]

    row_b_home = build_match_row(team_b, team_a, teams_data, X_columns)
    probs_b_home = model.predict_proba(row_b_home)[0]

    classes = model.classes_
    idx_win = list(classes).index(1)
    idx_lose = list(classes).index(0)

    p_a_wins = (probs_a_home[idx_win] + probs_b_home[idx_lose]) / 2
    p_b_wins = (probs_a_home[idx_lose] + probs_b_home[idx_win]) / 2

    total = p_a_wins + p_b_wins

    if np.isnan(total) or total == 0:
        p_a_wins, p_b_wins = 0.5, 0.5
    else:
        p_a_wins = p_a_wins / total
        p_a_wins = float(np.clip(round(p_a_wins, 10), 0.0, 1.0))
        p_b_wins = 1.0 - p_a_wins

    if deterministic:
        # ✅ خد الفريق صاحب أعلى احتمال مباشرة (ثابت دايمًا)
        winner = team_a if p_a_wins >= p_b_wins else team_b
    else:
        # الاختيار العشوائي (للـ Monte Carlo)
        winner = np.random.choice([team_a, team_b], p=[p_a_wins, p_b_wins])

    return winner
```

```python
round_of_32 = [

    # Match 73 & 74
    ('Germany', 'Paraguay'),
    ('France', 'Sweden'),

    # Match 75 & 76
    ('South Africa', 'Canada'),
    ('Netherlands', 'Morocco'),

    # Match 77 & 78
    ('Brazil', 'Japan'),
    ("Côte D'Ivoire", 'Norway'),

    # Match 79 & 80
    ('Mexico', 'Ecuador'),
    ('England', 'Congo Dr'),

    # Match 81 & 82
    ('Portugal', 'Croatia'),
    ('Spain', 'Austria'),

    # Match 83 & 84
    ('United States', 'Bosnia And Herzegovina'),
    ('Belgium', 'Senegal'),

    # Match 85 & 86
    ('Argentina', 'Cape Verde'),
    ('Australia', 'Egypt'),

    # Match 87 & 88
    ('Switzerland', 'Algeria'),
    ('Colombia', 'Ghana')
]

round_of_16_pairing = [
    (0,1),     # Match 89
    (2,3),     # Match 90
    (4,5),     # Match 91
    (6,7),     # Match 92
    (8,9),     # Match 93
    (10,11),   # Match 94
    (12,13),   # Match 95
    (14,15)    # Match 96
]

quarterfinal_pairing = [
    (0,1),     # Match 97
    (4,5),     # Match 98
    (2,3),     # Match 99
    (6,7)      # Match 100
]

semifinal_pairing = [
    (0,1),     # Match 101
    (2,3)      # Match 102
]

final_pairing = [
    (0,1)
]
```

```python
def build_next_round(winners, pairing):
    return [(winners[a], winners[b]) for a, b in pairing]
```

```python
def simulate_full_bracket(round_of_32, teams_data, model, X_columns):
    stages = [
        ("Round of 32", round_of_16_pairing),
        ("Round of 16", quarterfinal_pairing),
        ("Quarterfinal", semifinal_pairing),
        ("Semifinal", final_pairing),
        ("Final", None),
    ]

    current_matches = round_of_32
    all_results = {}
    winners = []

    for stage_name, pairing in stages:
        winners = []
        for team_a, team_b in current_matches:
            winner = simulate_knockout_match(team_a, team_b, teams_data, model, X_columns)
            winners.append(winner)

        all_results[stage_name] = list(zip(current_matches, winners))

        if pairing is not None:
            current_matches = build_next_round(winners, pairing)

    champion = winners[0]
    return champion, all_results
```
