#Data Cleaning

```python
# Handling nan values in columns ['pace','shooting','passing'] using KNN

from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

reference_cols = ['elo','overall','max_overall','attack','defense',
                   'pace','shooting','passing','form_scored',
                   'form_conceded','form_win_rate']

target_cols = ['pace','shooting','passing']

before_counts = data[target_cols].isnull().sum()

scaler = StandardScaler()
scaled_values = scaler.fit_transform(data[reference_cols])
scaled_data = pd.DataFrame(scaled_values, columns=reference_cols, index=data.index)

imputer = KNNImputer(n_neighbors=5, weights='distance')
imputed_scaled = imputer.fit_transform(scaled_data)

imputed_data = pd.DataFrame(
    scaler.inverse_transform(imputed_scaled),
    columns=reference_cols,
    index=data.index)

data[target_cols] = imputed_data[target_cols]

after_counts = data[target_cols].isnull().sum()
print("before:\n", before_counts)
print("\nAfter\n", after_counts)
```

```python
data['team'] = data['team'].str.strip().str.title()
```

```python
name_mapping = {
    'USA': 'United States',
    'United States Of America': 'United States',
    'South Korea': 'Korea Republic',
    'Korea Republic Of': 'Korea Republic',
    'North Korea': 'Korea DPR',
    'Ivory Coast': "Côte D'ivoire",
    'Cote Divoire': "Côte D'ivoire",
    'Dr Congo': 'Congo DR',
    'Democratic Republic Of Congo': 'Congo DR',
    'Congo Dr': 'Congo DR',
    'Republic Of Ireland': 'Ireland',
    'Uae': 'United Arab Emirates',
    'Uk': 'United Kingdom',
    'Ussr': 'Russia',
    'Czechia': 'Czech Republic',
    'Burma': 'Myanmar',
    'Swaziland': 'Eswatini',
    'Cape Verde': 'Cabo Verde',
    'East Timor': 'Timor-Leste',
    'Bosnia': 'Bosnia And Herzegovina',
    'Trinidad': 'Trinidad And Tobago',
    'St Kitts': 'Saint Kitts And Nevis',
}

data['team'] = data['team'].str.strip().str.title()

found_alternate_names = data[data['team'].isin(name_mapping.keys())]

print(f"number of teams that have different names: {len(found_alternate_names)}")
print(found_alternate_names[['team']])

data['team'] = data['team'].replace(name_mapping)

remaining = data[data['team'].isin(name_mapping.keys())]
print(f"\nAfter correction {len(remaining)}")
```
